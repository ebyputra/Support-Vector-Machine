```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_excel('klasi.xlsx')
df
```

```python
df.isna().sum()
```

```python
df.info() 
```

```python
N, D = df.shape
print('N: %d, D: %d' %(N, D))
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Pilek'] = encoder.fit_transform(df['Pilek'])
df['Demam'] = encoder.fit_transform(df['Demam'])
df['Batuk'] = encoder.fit_transform(df['Batuk'])
df['Sesak Nafas'] = encoder.fit_transform(df['Sesak Nafas'])
df['Keluhan Lain'] = encoder.fit_transform(df['Keluhan Lain'])
```

```python
df
```

```python
def ubah(v):
    if v == 1:
        return 0
    else:
        return 1
df['Pilek'] = df.Pilek.apply(ubah)

def ganti(w):
    if w == 1:
        return 0
    else:
        return 1
df['Gender'] = df.Gender.apply(ganti)
```

```python
df
```

```python
# Separating attributes
X = df[df.columns[:9]]

# Separating label
y = df['Status']
```

```python
from sklearn.preprocessing import StandardScaler

# standardization of data set values
scaler = StandardScaler()


scaler.fit(X)
X = scaler.transform(X)

X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2)
print(X_train.shape, y_train.shape)
```

```python
print(X_test.shape, y_test.shape)
```

```python
from sklearn import svm
from sklearn.svm import SVC

# create SVC object and call fit function to train model
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'kernel' : ['rbf', 'poly', 'sigmoid'],
    'C' : [0.5, 1, 10, 100],
    'gamma' : ['scale', 1, 0.1, 0.01, 0.001]
}

grid_search = GridSearchCV(estimator=SVC(random_state=0),
                          param_grid=parameters,
                          n_jobs=6,
                          verbose=1,
                          scoring='accuracy')

grid_search.fit(X_train, y_train)
```

```python
print(f'Best Score: {grid_search.best_score_}')

best_params = grid_search.best_estimator_.get_params()
print(f'Best Parameters:')
for param in parameters:
    print(f'\t{param}: {best_params[param]}')
```

```python
y_pred = grid_search.predict(X_test)

print(classification_report(y_test, y_pred))
```
